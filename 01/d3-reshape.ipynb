```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv", converters={'Randomized Id': str})
df
```

```python
df["unique_id"] = df["UniqueRandomComplaintId"].astype(str) + "-" + df["Allegation Id"].astype(str)
```

```python
reshaped = pd.DataFrame(df.groupby(["Allegation FADO Type", "Allegation Disposition"])["unique_id"].count().unstack())
reshaped
```

```python
reshaped = reshaped.fillna(0)
reshaped
```

```python
reshaped
```

```python
reshaped['Total']= reshaped.sum(axis=1)
reshaped
```

```python
reshaped.to_csv("reshaped.csv")
```
