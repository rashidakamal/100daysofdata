```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv", converters={'Randomized Id': str})
df
```

```python
 df["unique_id"] = df["UniqueRandomComplaintId"].astype(str) + "-" + df["Allegation Id"].astype(str)
```

```python
total_by_year = pd.DataFrame(df['Closed Year'].value_counts().sort_index())
total_by_year
```

```python
by_year_calc = 600 / total_by_year["Closed Year"].max()
by_year_calc
```

```python
total_by_year["size"] = total_by_year["Closed Year"] * by_year_calc
```
